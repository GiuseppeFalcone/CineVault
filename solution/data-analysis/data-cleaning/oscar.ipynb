# Oscars Dataset Cleaning
This notebook cleans and prepares the oscar awards data for further data analysis

```python
import pandas as pd
```

```python
oscar_df = pd.read_csv('../../data/the_oscar_awards.csv')
oscar_df.info()
print(oscar_df.head())
```

## Dataset Overview
The Dataset contains information about actors who have won an oscar.
The data is divided in **6 columns**:
- **Year film**: Represents the year in which the Oscar-nominated film was released.
- **Year ceremony**: Represents the year in which the Oscar ceremony was held.
- **Ceremony**: How many times the nominee was nominated for an Oscar Award.
- **Category**: The role for which the nomination is for.
- **Name**: The name who was nominate for the Oscar Award.
- **Film**: The name of the film nominate for the Oscar Award.
- **Winner**: If the nominee noominated for an Oscar Award won the Oscar or not.

---

## Starting the data cleaning process
The goals are:
- Looking for null values
- Looking for duplicates and redundant data
- Changing column entries values
- Casting column types to appropriate dtypes

#### 1. Looking for null values

```python
null_values = oscar_df.isnull().sum()
print(null_values)
```

#### 2. Looking for duplicates and redundant data
Controllo se ci sono righe duplicate e le rimuovo.

```python
duplicates = oscar_df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
duplicates_rows = oscar_df[oscar_df.duplicated(keep=False)]
oscar_df=oscar_df.drop_duplicates(keep='first').reset_index(drop=True)
```

#### 3. Changing 'category' column entries values
trasformo le stringhe in carattere minuscolo

```python
oscar_df['category'] = oscar_df['category'].str.lower()
```

#### 4. Casting column types to appropriate dtypes
Cambio i tipi dei valori delle tabelle.

```python
oscar_df['category'] = oscar_df['category'].astype('category')
oscar_df['name'] = oscar_df['name'].astype('category')
oscar_df['film'] = oscar_df['film'].astype('category')
oscar_df['winner'] = oscar_df['winner'].astype(bool)
```

### Saving the cleaned DataFrame to a new CSV file

```python
oscar_df.to_csv('../../data/data-cleaned/oscars_cleaned.csv', index=False)
print("DataFrame saved to CSV file.")
```
