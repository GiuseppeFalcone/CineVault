# Crew Dataset Cleaning
This notebook cleans and prepares the crew data for further data analysis

```python
import pandas as pd
```

```python
crew_df = pd.read_csv('../../data/crew.csv')
crew_df.info()
```

## Dataset Overview
The Dataset contains information about the crew.
The data is divided in **3 columns**:
- **Id**: A unique identifier for each movie.
- **Role**: The role of a crew member.
- **name**: The name of a crew member.

---

#### 1. Check for null values

```python
null_counts=crew_df.isnull().sum()
print(null_counts)
```

#### 2. Identify and remove duplicate rows, keeping first occurrence
Calcolo duplicati e rimuovo mantenendo il primo.


```python
duplicates=crew_df.duplicated().sum()
print(f"Number of duplicates rows: {duplicates}")
duplicates_row = crew_df[crew_df.duplicated(keep=False)]
crew_df = crew_df.drop_duplicates(keep='first').reset_index(drop=True)
```

#### 3. Cast to category
Cambiando il tipo dei valori

```python
crew_df['role']=crew_df['role'].astype('category')
crew_df['name']=crew_df['name'].astype('category')
```

#### 3. Rename the crew columns
Cambio nome alle colonne

```python
crew_df.rename(columns = {'name':'crew_name', "role":"crew_role"}, inplace = True)
```

### Saving the cleaned Data

```python
crew_df.to_csv('../../data/data-cleaned/crew_cleaned.csv', index=False)
print("Saved the cleaned data to CSV")
```
