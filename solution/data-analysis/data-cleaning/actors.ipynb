# Actors Dataset Cleaning
This notebook cleans and prepares the actors data for further data analysis

```python
import pandas as pd
```

```python
actors_df = pd.read_csv('../../data/actors.csv')
print(actors_df.info())
print(actors_df.head())
```

## Dataset Overview
The Dataset contains information about the studios.
The data is divided in **2 columns**:
- **Id**: A unique identifier for each movie.
- **Name**: The name of actor.
- **Role**: What role he played.

---

#### 1. Check for null values

```python
null_counts = actors_df.isnull().sum()
print(null_counts)
```

#### 2. Identify and remove duplicate rows, keeping first occurrence
Calcolo duplicati e rimuovo mantenendo il primo.


```python
duplicates=actors_df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
duplicates_row = actors_df[actors_df.duplicated(keep = False)]
actors_df=actors_df.drop_duplicates(keep = 'first').reset_index(drop = True)
```

#### 3. Rename the actors columns
Cambio nome alle colonn di actors

```python
actors_df.rename(columns = {'name':'actor_name', "role":"actor_role"}, inplace = True)
```

#### 4. Cast to category
Cambiando il tipo dei valori

```python
actors_df['actor_name']=actors_df['actor_name'].astype('category')
actors_df['actor_role']=actors_df['actor_role'].astype('category')
```

### Saving the cleaned Data to a new CSV File

```python
print(actors_df.info())
```

```python
actors_df.to_csv('../../data/data-cleaned/actors_cleaned.csv', index = False)
print("Saved the cleaned CSV")
```
