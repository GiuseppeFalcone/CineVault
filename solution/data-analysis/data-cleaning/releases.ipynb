# Releases Dataset Cleaning
This notebook cleans and prepares the releases data for further data analysis

```python
import pandas as pd
```

```python
releases_df = pd.read_csv('../../data/releases.csv')
print(releases_df.info())
print(releases_df.head())
```

## Dataset Overview
The Dataset contains information about the releases.
The data is divided in **5 columns**:
- **Id**: A unique identifier for each movie.
- **Country**: Indicates where the movie was released.
- **date**: Indicates when the movie was released.
- **type**:Indicates how the movie was released.
- **rating**: The rating assigned to a film by each regional authority for its suitability for viewing, for example PG, PEGI 18, and the like.

---

#### 1. Check for null values

```python
null_counts = releases_df.isnull().sum()
print(null_counts)
```

#### 2. Identify and remove duplicate rows, keeping first occurrence
Calcolo duplicati e rimuovo mantenendo il primo.


```python
duplicates = releases_df.duplicated().sum()
print(f"Number of duplicate values: {duplicates}")
```

#### 3. Rename the releases's column
Cambio nome alla colonna di releases

```python
releases_df.rename(columns={
    'rating':'release_age_rating',
    'type': 'release_type',
    'date': 'release_date',
    'country': 'release_country'},
inplace=True)
```

#### 4. Cast to category
Cambiando il tipo dei valori

```python
releases_df['release_date'] = pd.to_datetime(releases_df['release_date'],errors='coerce')
releases_df['release_country']=releases_df['release_country'].astype('category')
releases_df['release_type']=releases_df['release_type'].astype('category')
releases_df['release_age_rating']=releases_df['release_age_rating'].astype('category')
```

### Saving the cleaned DataFrame to a new CSV file

```python
releases_df.to_csv('../../data/data-cleaned/releases_cleaned.csv', index=False)
print("Cleaned DataFrame saved to 'releases_cleaned.csv'")
```
