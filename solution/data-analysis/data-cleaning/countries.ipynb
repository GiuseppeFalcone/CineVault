# Countries Data Cleaning
This notebook cleans and prepares the countries data for further data analysis.

```python
import pandas as pd
```

```python
countries_df = pd.read_csv('../../data/countries.csv')
print(countries_df.info())
print(countries_df.head())
```

## Dataset Overview
The Dataset contains information about movies production countries.
The data is divided in **2 columns**:
- **id**: A unique identifier for each movie
- **country**: Provides the value for each country where the movie was produced.
---
## Starting the data cleaning process
The goals are:
- **Handling duplicates and redundant data**: since ids have to be unique for each movie
- **Handling logical inconsistency**
- **Casting *country* column values to *category***
- **Rename column**: to give more unique identifiable meaning
- **Look for wrong values**: we have to be sure that the dataset does not contain outliers, wrong or unrealistic values.

### 1. Looking for duplicates

```python
print(f"Number of duplicate (id, country) pairs: {countries_df.duplicated(subset=['id', 'country']).sum()}")
```

### 2. Casting *country* column values to *category*

```python
countries_df['country'] = countries_df['country'].astype('category')
```

### 3. Renaming country column to *production_country*

```python
countries_df.rename(columns={'country': 'production_country'}, inplace=True)
```

### 4. Looking for redundant values

```python
print("Unique countries before cleaning:")
print(countries_df['production_country'].unique())

# Strip leading/trailing whitespace
countries_df['production_country'] = countries_df['production_country'].str.strip()

# Creating a mapping dictionary for common country name variations
country_mapping = {
    'US': 'United States of America',
    'USA': 'United States of America',
    'U.S.A.': 'United States of America',
    'United States': 'United States of America',
    'UK': 'United Kingdom',
    'U.K.': 'United Kingdom',
    'Soviet Union': 'Russia',
    'FR': 'France',
    'DE': 'Germany',
    'JP': 'Japan',
    'CN': 'China',
    'CA': 'Canada',
    'AU': 'Australia',
    'IT': 'Italy',
    'ES': 'Spain',
    'BR': 'Brazil',
    'IN': 'India',
    'ZA': 'South Africa',
    'The Netherlands': 'Netherlands',
    'Holland': 'Netherlands',
    'Czechia': 'Czech Republic'
}

# Apply the mapping to the 'production_country' column
countries_df['production_country'] = countries_df['production_country'].replace(country_mapping)

# Update the 'category' type after string manipulations
countries_df['production_country'] = countries_df['production_country'].astype('category')

print("\nUnique countries after cleaning:")
print(countries_df['production_country'].unique())
```

### Saving the cleaned dataset

```python
print(countries_df.info())
```

```python
countries_df.to_csv('../../data/data-cleaned/countries_cleaned.csv', index=False)
print("Saved the cleaned DataFrame to CSV")
%reset -f
```
