# Genres Dataset Cleaning
This notebook cleans and prepares the genres data for further data analysis.

```python
import pandas as pd
```

```python
genres_df = pd.read_csv('../../data/genres.csv')
print(genres_df.info())
print(genres_df.head())
```

## Dataset Overview
The Dataset contains information about genres of each movie.
The data is divided in **2 columns**:
- **id**: A unique identifier for each movie.
- **genre**: Provides a genre for the movie.

---
## Starting the data cleaning process
The goals are:
- **Handling duplicates and redundant data**: since ids have to be unique for each movie
- **Casting *genre* column values to *category***
- **Rename columns**: to give more unique identifiable meaning

### 1. Looking for duplicates and redundant data
   1. Checking if the tuple ('id', 'genre') is duplicated

```python
print(f"Number of duplicate (id, genre) pairs: {genres_df.duplicated(subset=['id', 'genre']).sum()}")
```

2. Checking and handling **Inconsistent** or **Redundant Data**

```python
print("Unique genres before cleaning:")
print(genres_df['genre'].unique())
```

There are no wrong values in the *genre* column.
### 2. Renaming *genre* column and casting to *category*

```python
genres_df['genre'] = genres_df['genre'].astype('category')
genres_df.rename(columns={'genre': 'genre_category'}, inplace=True)
print(genres_df.info())
```

### Saving the cleaned dataset

```python
genres_df.to_csv('../../data/data-cleaned/genres_cleaned.csv', index=False)
print("\nCleaned genres DataFrame saved to '../../data/data-cleaned/genres_cleaned.csv'")
```
