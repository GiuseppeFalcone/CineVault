# Themes Dataset Cleaning
This notebook cleans and prepares the themes data for further data analysis

```python
import pandas as pd
```

```python
themes_df = pd.read_csv('../../data/themes.csv')
print(themes_df.info())
print(themes_df.head())
```

## Dataset Overview
The Dataset contains information about the studios.
The data is divided in **2 columns**:
- **Id**: A unique identifier for each movie.
- **Theme**: The theme for each movie.

---

#### 1. Check for null values

```python
null_counts = themes_df.isnull().sum()
print("Null values for column:")
print(null_counts)
```

#### 2. Identify and remove duplicate rows, keeping first occurrence
Calcolo duplicati e rimuovo mantenendo il primo.

```python
duplicates = themes_df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

#### 3. Cast to category
Cambiando il tipo dei valori

```python
themes_df['theme'] = themes_df['theme'].astype('category')
```

### Saving the cleaned DataFrame to a new CSV file

```python
themes_df.to_csv('../../data/data-cleaned/themes_cleaned.csv', index=False)
print("Cleaned DataFrame saved to 'posters_cleaned.csv'")
```
