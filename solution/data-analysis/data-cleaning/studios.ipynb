# Studios Dataset Cleaning
This notebook cleans and prepares the studios data for further data analysis

```python
import pandas as pd
```

```python
studios_df = pd.read_csv('../../data/studios.csv')
studios_df.info()
```

## Dataset Overview
The Dataset contains information about the studios.
The data is divided in **2 columns**:
- **Id**: A unique identifier for each movie.
- **Studio**: The name of studio.

---

#### 1. Check for null values

```python
null_counts = studios_df.isnull().sum()
print("Null values for column:")
print(null_counts)
```

#### 2. Identify and remove duplicate rows, keeping first occurrence
Calcolo duplicati e rimuovo mantenendo il primo.


```python
duplicates = studios_df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
duplicates_row = studios_df[studios_df.duplicated(keep = False)]
studios_df=studios_df.drop_duplicates(keep = 'first').reset_index(drop = True)
```

#### 3. Rename the studio's column
Cambio nome alla colonna di studio

```python
studios_df.rename(columns={'studio': 'production_studio'}, inplace=True)
```

#### 4. Cast to category
Cambiando il tipo dei valori

```python
studios_df['production_studio'] = studios_df['production_studio'].astype('category')
```

### Saving the cleaned DataFrame to a new CSV file

```python
studios_df.to_csv('../../data/data-cleaned/studios_cleaned.csv', index=False)
print("Cleaned DataFrame saved to 'studios_cleaned.csv'")
```
