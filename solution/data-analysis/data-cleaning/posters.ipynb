# Posters Data Cleaning
This notebook cleans and prepares the posters data for further data analysis.

```python
import pandas as pd
```

```python
posters_df = pd.read_csv('../../data/posters.csv') # Make sure to replace with the actual path
print(posters_df.info())
print(posters_df.head())
```

## Dataset Overview
The Dataset contains information about movies posters.
The data is divided in **2 columns**:
- **id**: A unique identifier for each movie.
- **link**: Provides the letterbox link to the movie poster.
---

## Starting the data cleaning process
The goals are:
- **Handling duplicates and redundant data**: since ids have to be unique for each movie
- **Casting *link* column values to *string***
- **Rename columns**: to give more unique identifiable meaning

### 1. Looking for duplicates and redundant values

```python
id_duplicates = posters_df.duplicated(subset=['id'], keep=False)
num_id_duplicates = id_duplicates.sum()
print(f"Number of duplicated rows based on 'id': {num_id_duplicates}")
```

There are **no** fully duplicated rows based on 'id'.

### 2. Casting link column values to string

```python
posters_df['link'] = posters_df['link'].astype(str)
```

### 3. Renaming link column

```python
posters_df.rename(columns={'link': 'poster_link'}, inplace=True)
```

### Checking if everything went ok

```python
print(posters_df.info())
print(posters_df.head())
```

### Saving the cleaned DataFrame to a new CSV file

```python
posters_df.to_csv('../../data/data-cleaned/posters_cleaned.csv', index=False)
print("Cleaned DataFrame saved to 'posters_cleaned.csv'")
```
